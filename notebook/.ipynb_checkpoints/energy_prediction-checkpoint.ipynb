## Data Preprocessing

```python
df.replace(['???', 'na', 'NA', 'null', '', 'None'], np.nan, inplace=True)

# Now convert to numeric
for col in df.columns:
    if col != 'timestamp':
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df_processed = df.copy()
X, transformers = preprocess_data(df_processed, is_training=True)
y = np.log1p(df_processed['equipment_energy_consumption'])

# Drop rows where y is NaN or infinite
valid_indices = ~np.isnan(y) & ~np.isinf(y)
X = X[valid_indices].reset_index(drop=True)
y = y[valid_indices].reset_index(drop=True)
print(f'Dropped {len(df) - len(y)} rows with NaN/infinite values in y.')
```

## Feature Selection

```python
# Add random variables back to X for evaluation
df_temp = df.copy()
df_temp = df_temp[valid_indices]  
X_with_random = X.copy()
X_with_random['random_variable1'] = df_temp['random_variable1'].values
X_with_random['random_variable2'] = df_temp['random_variable2'].values

# Ensure random variables are numeric
X_with_random['random_variable1'] = pd.to_numeric(X_with_random['random_variable1'], errors='coerce')
X_with_random['random_variable2'] = pd.to_numeric(X_with_random['random_variable2'], errors='coerce')

# Impute and scale the random variables
imputer = transformers['imputer']
scaler = transformers['scaler']
random_vars = pd.DataFrame(imputer.fit_transform(X_with_random[['random_variable1', 'random_variable2']]), 
                          columns=['random_variable1', 'random_variable2'])
random_vars = pd.DataFrame(scaler.fit_transform(random_vars), columns=['random_variable1', 'random_variable2'])
X_with_random[['random_variable1', 'random_variable2']] = random_vars

# Train a temporary model 
temp_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
temp_model.fit(X_with_random, y)

# Feature importance plot
importances = pd.Series(temp_model.feature_importances_, index=X_with_random.columns)
importances.sort_values().plot(kind='barh', figsize=(10, 8))
plt.title('Feature Importance')
plt.savefig('../reports/figures/feature_importance.png')
plt.show()

# Correlation with target
df_temp['log_equipment_energy'] = y
random_corr = df_temp[['log_equipment_energy', 'random_variable1', 'random_variable2']].corr()
print('Correlation with log-transformed target:')
print(random_corr)
```

## Model Development

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# XGBoost model
model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)

# Cross-validation
cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
cv_rmse = np.sqrt(-cv_scores)
print(f'Cross-Validation RMSE Scores: {cv_rmse}')
print(f'Average CV RMSE: {cv_rmse.mean():.2f} (+/- {cv_rmse.std() * 2:.2f})')

# Fit the model on the full training data
model.fit(X_train, y_train)
```

## Model Evaluation

```python
# Predictions
y_pred = predict_energy(model, X_test)
y_test_orig = np.expm1(y_test)

# Metrics
rmse = np.sqrt(mean_squared_error(y_test_orig, y_pred))
mae = mean_absolute_error(y_test_orig, y_pred)
r2 = r2_score(y_test_orig, y_pred)
print(f'RMSE: {rmse:.2f} Wh')
print(f'MAE: {mae:.2f} Wh')
print(f'RÂ²: {r2:.2f}')

# Plot predictions vs actual
plt.figure(figsize=(10, 6))
plt.scatter(y_test_orig, y_pred, alpha=0.3)
plt.plot([y_test_orig.min(), y_test_orig.max()], [y_test_orig.min(), y_test_orig.max()], 'r--')
plt.title('Predicted vs Actual Energy Consumption')
plt.xlabel('Actual (Wh)')
plt.ylabel('Predicted (Wh)')
plt.savefig('../reports/figures/pred_vs_actual.png')
plt.show()
```
